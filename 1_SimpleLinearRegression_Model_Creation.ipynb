```python
# pandas is library class/function which can handle Databases and Tables
import pandas as p
```

```python
# upload csv file-salary_data.csv into variable dataset by using pandas.read_csv function
dataset = p.read_csv("Salary_Data.csv")

# view the table_data through variable dataset
dataset
```

```python
# Split Input data from above data-table and load it to variable independent
independent=dataset[["YearsExperience"]]

# view the input data from variable independent 
independent
```

```python
# Split Output data from above data-table and load it to variable as dependent
dependent=dataset[["Salary"]]

# view the output-data from variable dependent
dependent
```

```python
# Training_set and Test_Set splitting from input data(independent) & output data(dependent) from SKlearn libraries
# Considering memory consumption calling model_selection function for selecting model 
# Spliting train_set & test_set using train_test_split function 

from sklearn.model_selection import train_test_split
```

```python
# providing input parameters as independent, output parameters as dependent
# giving training_set data size as 70% & test_set data size as 30% of input and output dataset
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# instead of getting all combined values in a single dataset, based on SKLearn splitting the values as 
# Training_input = x_train, Test_input = x_test
# Training_Output = y_train, Test_output = y_test

x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# view training_set values from the input data Independent
x_train
```

```python
# view test_set values from the input data Independent
x_test
```

```python
# view training_set values from the Output data Dependent
y_train
```

```python
# view test_set values from the Output data Dependent
y_test
```

```python
# based on SKLearn Library class/function, calling linear_model function & LinearRegression()

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
# viewing the regressor procedure values
regressor
```

```python
# substituting train_set values by providing input parameter as x_train & output parameter as y_train by using fit()
regressor.fit(x_train,y_train)
```

```python
# to view slope/weight by using coef_
weight=regressor.coef_
weight # view the output of weight
```

```python
# to view bias/intercept by using intercept_
bais=regressor.intercept_
bais # view the output of bais
```

```python
# Evaluating Metrics by checking model efficiency for test_set values by using predict()
y_pred=regressor.predict(x_test)
y_pred # view the output of y_pred
```

```python
# Evaluating Metrics values by using r2_score
from sklearn.metrics import r2_score
r2_score # view the output of r2_score
```

```python
# Evaluating/comparing output values (y_test) and predicted values(y_pred, 
# if result of r_score is near to 1-good Model or near to 0-bad Model
r_score=r2_score(y_test,y_pred)
r_score # view the output of r_score
```

```python
# saving the model by using pickle class/function

import pickle
filename="LinearRegressionModel1_final.sav" # creating dummy filename and saving the model
```

```python
# dumping the model from regressor by using pickle dump(), open() with write-binary option
pickle.dump(regressor,open(filename,"wb"))
```

```python
# load and check the Model whether it's working properly by using pickle load(),open() with read-binary option
loaded_model=pickle.load(open("LinearRegressionModel1_final",'rb'))
result=loaded_model.predict([[13]]) # check and confirm the predicted output result by inputting values
result # view the output of result 
```

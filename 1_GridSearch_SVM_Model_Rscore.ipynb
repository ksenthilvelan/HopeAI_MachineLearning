```python
import pandas as p
```

```python
dataset=p.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=p.get_dummies(dataset)
dataset
```

```python
dataset=p.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
# Input dataset

independent = dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
# Output dataset

dependent = dataset[['charges']]
dependent
```

```python
# using SKLearn libraries model_selection, spliting input and output values for training-set values & test-set values

from sklearn.model_selection import train_test_split
```

```python
# providing input parameters as independent, output parameters as dependent
# giving training_set data size as 70% & test_set data size as 30% of input and output dataset

train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# instead of getting all combined values in a single dataset, based on SKLearn splitting the values as 
# Training_input = x_train, Test_input = x_test and Training_Output = y_train, Test_output = y_test

x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)

x_train # to view training - input values
```

```python
x_test # to view test - input values
```

```python
y_train # to view training - output values
```

```python
y_test # to view test - output values
```

```python
# Applying Support Vector Machine Algorithm 
# based on SKLearn preprocessing Libraries, using StandardScaler()

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# fit_tranform() calculates x_train values & transform to evaluate mean and std deviation output values
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# based on SKLearn svm libraries, using GridSearchCV(),SVR()

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
              'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}

grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3, n_jobs = -1)

# fitting the model for grid search

grid.fit(x_train,y_train)
```

```python
# # based on SKLearn svm libraries, using SVR()
# from sklearn.svm import SVR
# regressor = SVR(kernel='rbf',C=10000)
# regressor
```

```python
# print best parameter after tuning
# print(grid.best_params_)
re = grid.cv_results_

print("The R-Score value for the best parameter {} :".format(grid.best_params_))
```

```python
# View in Table form

table = p.DataFrame.from_dict(re)
table
```

```python
# Check and verify

age_input=float(input("Age :"))
bmi_input=float(input("BMI :"))
children_input=float(input("Children :"))
sex_female_input=int(input("Sex: If female give value as 1 otherwise 0 -"))
sex_male_input=int(input("Sex: If male give value as 1 otherwise 0 -"))
smoker_no_input=int(input("Smoker: If non-smoker give value as 1 otherwise 0 -"))
smoker_yes_input=int(input("Smoker:If smoker give value as 1 otherwise 0 -"))
```

```python
Future_Prediction = grid.predict([[age_input,bmi_input,children_input,sex_female_input,sex_male_input,smoker_no_input,smoker_yes_input]])
# Future_Prediction
print("Future_Prediction={}".format(Future_Prediction))
```
